```python
import pandas as pd
import numpy as np 
#from test_case_generator_functions import art
```

```python

uid = 'UID-110' 
theme = 'Different Name Parts'
category = 'Articles'
sub_category = '> 2 Articles difference '
entity_type = 'Entity'
```

```python
ofac_list_download = pd.read_csv('https://www.treasury.gov/ofac/downloads/sdn.csv', header=None)
ofac_list_download
```

```python
ofac_list = ofac_list_download[[0,1,2]]
ofac_list.columns = ['uid', 'name', 'entity_type']
ofac_list
```

```python
ofac_list['entity_type'].value_counts()
```

```python
ofac_list_filtered = ofac_list[(ofac_list.entity_type == '-0- ')] # only evaluate Entity
ofac_list_filtered
```

```python
ofac_list_sampled = ofac_list_filtered.sample(n = 10)
ofac_list_sampled
```

```python
final_test_cases = pd.DataFrame(columns=['UID', 'Theme','Category','Sub-category','Entity-Type','Test Case ID' , 'OFAC List UID', 'Original Name','Test Case Name'])
final_test_cases
```

```python
def art(x):
    import numpy as np
    articles = ['A', 'AN', 'THE', 'AND', 'OF', 'OR', 'IN', 'ON', 'AT']
    if x in articles:
        articles.remove(x)
        return np.random.choice(articles)
    else:
        return x
```

```python
for index, row in ofac_list_sampled.iterrows():
    original_name = row['name']
    
    split_name = original_name.split(' ')
    for i in range(len(split_name)):
        split_name[i] = art(split_name[i])
    final_test_name = ' '.join(split_name)
    
    final_test_cases.loc[len(final_test_cases)] = [uid, theme, category, sub_category, entity_type, uid + ' - ' + str(index), row['uid'], row['name'], final_test_name] # append to the dataframe

final_test_cases
```
