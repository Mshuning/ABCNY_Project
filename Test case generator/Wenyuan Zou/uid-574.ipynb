```python
# -----------------------------------
# TEST CASE TYPE REFERENCE INFORMATION
# -----------------------------------

uid = 'UID-574' 
theme = 'Names where Name Parts are Modified'
category = 'Transpositions'
sub_category = '2 transpositions (same name part) '
entity_type = 'Vessel'

# ********************
```

```python
import random
def transport_anywhere(phrase,k):
    for num in random.sample(range(0, len(phrase)-1, k),k):
        if phrase[num+1] != phrase[num]:
            phrase = phrase[:num] +phrase[num+1]+phrase[num]+ phrase[num+2:]
        else:
            phrase = phrase[:num+1] +phrase[num+2]+phrase[num+1]+ phrase[num+3:]
    return phrase
```

```python
import pandas as pd
from test_case_generator_functions import alphabet

# download the OFAC list from the web

ofac_list_download = pd.read_csv('https://www.treasury.gov/ofac/downloads/sdn.csv', header=None)
ofac_list = ofac_list_download[[0,1,2]]
ofac_list.columns = ['uid', 'name', 'entity_type']
```

```python
# new data frame with split value columns
ofac_split = ofac_list["name"].str.split(" ", n = 1, expand = True)



# making separate first name column from new data frame
ofac_list["First Name"]= ofac_split[0]
 
# making separate last name column from new data frame
ofac_list["Last Name"]= ofac_split[1]
 
ofac_list_filtered = ofac_list[(ofac_list.entity_type == 'vessel')& (ofac_list['First Name'].str.len()>5)] # only evaluate vessel

#ofac_list_filtered = ofac_list_filtered[(ofac_list_filtered.name.str.count(' ') >0)]
ofac_list_sampled = ofac_list_filtered.sample(n = 10)
ofac_list_sampled
```

```python
# create blank final test cases table

final_test_cases = pd.DataFrame(columns=['UID', 'Theme','Category','Sub-category','Entity-Type','Test Case ID' , 'OFAC List UID', 'Original Name','Test Case Name'])
final_test_cases
```

```python
# run loop to generate the test cases

for index, row in ofac_list_sampled.iterrows():
        number_of_words = row['name'].count(' ') + 1 # determine for each name the number of words
        name_words=row['name'].split()
        if number_of_words==1:
            final_test_name=transport_anywhere(name_words[0],2)
            final_test_cases.loc[len(final_test_cases)] = [uid, theme, category, sub_category, entity_type, uid + ' - ' + str(index), row['uid'], row['name'], final_test_name] # append to the dataframe
        elif number_of_words==2 :
            final_test_name=transport_anywhere(name_words[0],2)+' '+name_words[1]
            final_test_cases.loc[len(final_test_cases)] = [uid, theme, category, sub_category, entity_type, uid + ' - ' + str(index), row['uid'], row['name'], final_test_name] # append to the dataframe
        elif number_of_words==3 :
            final_test_name=transport_anywhere(name_words[0],2)+' '+name_words[1]+' '+name_words[2]
            final_test_cases.loc[len(final_test_cases)] = [uid, theme, category, sub_category, entity_type, uid + ' - ' + str(index), row['uid'], row['name'], final_test_name] # append to the dataframe
        else:
            name_words=row['name'].split()
            final_test_name_0=''
            for i in range(2,len(name_words)):
                final_test_name_0=final_test_name_0+' '+name_words[i]
            final_test_name=transport_anywhere(name_words[0],2)+' '+final_test_name_0# create the test case
            final_test_cases.loc[len(final_test_cases)] = [uid, theme, category, sub_category, entity_type, uid + ' - ' + str(index), row['uid'], row['name'], final_test_name] # append to the dataframe

            

final_test_cases      

final_test_cases      
```
