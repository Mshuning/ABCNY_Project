```python
# -----------------------------------
# TEST CASE TYPE REFERENCE INFORMATION
# -----------------------------------

uid = 'UID-625' 
theme = 'Special Characters and Spaces'
category = 'Different Character'
sub_category = 'All special characters replaced by equivalent'
entity_type = 'Vessel'

# ********************
```

```python
def removal_chara(phrase):
    if '.' in phrase :
        index=phrase.find('.')
        phrase = phrase[:index]  +','+ phrase[index+1:]
    elif '-' in phrase :
        index=phrase.find('-')
        phrase = phrase[:index]  +'_'+ phrase[index+1:]
    else:
        phrase=phrase
    return phrase
```

```python
removal_chara('S.KUZNETSOV')
```

```python
import pandas as pd

# download the OFAC list from the web

ofac_list_download = pd.read_csv('https://www.treasury.gov/ofac/downloads/sdn.csv', header=None)
ofac_list = ofac_list_download[[0,1,2]]
ofac_list.columns = ['uid', 'name', 'entity_type']
```

```python
ofac_list = ofac_list_download[[0,1,2]]
ofac_list.columns = ['uid', 'name', 'entity_type']
```

```python
ofac_list_filtered =ofac_list [(ofac_list.entity_type == 'vessel')]
pat = '[(-.:/,#%\=$^&*@)]'
#special_characters = ['@','#','$','%','^', '&','*','(',')','()']
ofac_list_filtered['count'] = ofac_list_filtered['name'].str.count(pat)
ofac_list_filtered
test=ofac_list_filtered[ofac_list_filtered['count'] > 0 ]
test
```

```python
ofac_list_sampled = test.sample(n = 10)

ofac_list_sampled
```

```python
# create blank final test cases table

final_test_cases = pd.DataFrame(columns=['UID', 'Theme','Category','Sub-category','Entity-Type','Test Case ID' , 'OFAC List UID', 'Original Name','Test Case Name'])
final_test_cases
```

```python
for index, row in ofac_list_sampled.iterrows():
        number_of_words = row['name'].count(' ') + 1 # determine for each name the number of words
        name_words=row['name'].split()
        if number_of_words==1:
            final_test_name=removal_chara(name_words[0])
        else:
            final_test_name=removal_chara(name_words[0])+' '+removal_chara(name_words[1])
        final_test_cases.loc[len(final_test_cases)] = [uid, theme, category, sub_category, entity_type, uid + ' - ' + str(index), row['uid'], row['name'], final_test_name] # append to the dataframe
final_test_cases
```
