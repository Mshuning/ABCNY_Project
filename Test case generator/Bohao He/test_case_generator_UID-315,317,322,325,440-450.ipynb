```python
def month_dict():
    dic = {'Jan':'01','Feb':'02','Mar':'03',
           'Apr':'04','May':'05','Jun':'06',
           'Jul':'07','Aug':'08','Sep':'09',
           'Oct':'10','Nov':'11','Dec':'12'}
    return dic
```

```python
ofac_list_dob_filtered = ofac_list_dob.copy()
#extract date of birth
ofac_list_dob_filtered['DOB']=ofac_list_dob_filtered['DOB'].apply(lambda x : (x.split(';')[0])[4:])
ofac_list_dob_filtered['DOB']=ofac_list_dob_filtered['DOB'].apply(lambda x : (x.split('to')[0]) if re.search('to',x) else x)
for i in month_dict().keys():
    ofac_list_dob_filtered['DOB']=ofac_list_dob_filtered['DOB'].apply(lambda x : re.sub(i,month_dict()[i],x) if re.search(i,x) else x)
ofac_list_dob_filtered['DOB']=ofac_list_dob_filtered['DOB'].apply(lambda x : (x.split('circa')[1]) if re.search('circa',x) else x)
```
